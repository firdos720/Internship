Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and 
salary filter.  

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.naukri.com/')
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Delhi/NCR')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
job_loc=[]
company=[]
exp=[]
salary=[]
```

```python
title_tags=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
     job_title.append(i.text) 
loc_tags=driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
                                
for i in loc_tags:
    job_loc.append(i.text)
    
com_tags=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')  
for i in com_tags:
    company.append(i.text)
    
exp_tags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in exp_tags:
    exp.append(i.text)
sal_tags=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal"]')
for i in sal_tags:
    salary.append(i.text)
    
    
```

```python
df=pd.DataFrame({'Job_Title':job_title,'Location':job_loc,'company':company,'Experience':exp,'salary':salary})
df[0:11]
```
